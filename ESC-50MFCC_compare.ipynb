```python
# Imports
import numpy as np
import librosa
import matplotlib.pyplot as plt
```

```python
npz_file_path = "/content/drive/My Drive/Summer 2025/ESC-50 Water Project/audio/npz_files/esc50_mfcc_n64_fft2048_ovlp50_seg1000ms_natural.npz"
```

```python
# load data from .npz file
data = np.load(npz_file_path, allow_pickle=True)
mfccs = data['mfcc']
labels = data['labels']
mapping = data['mapping']
```

```python
# Find indices for the categories to comapre
pouring_water_label = np.where(mapping == 'pouring_water')[0][0]
water_drops_label = np.where(mapping == 'water_drops')[0][0]

# Filter MFCCs for each category
pouring_water_mfccs = mfccs[labels == pouring_water_label]
water_drops_mfccs = mfccs[labels == water_drops_label]

# Choose an example from each for visualization
pouring_water_example = pouring_water_mfccs[0]
water_drops_example = water_drops_mfccs[0]

# Plot the MFCCs side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

librosa.display.specshow(pouring_water_example, sr=22050, x_axis='time', ax=ax1)
ax1.set(title='MFCCs for Pouring water')

librosa.display.specshow(water_drops_example, sr=22050, x_axis='time', ax=ax2)
ax2.set(title='MFCCs for Water Drops')

plt.tight_layout()
plt.show()
```
